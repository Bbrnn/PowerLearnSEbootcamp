Task 1: Load and Explore the Dataset

```python
#Task 1: Load and Explore the Dataset
#IRIS DATASET
import pandas as pd
#load from sklearn
from sklearn.datasets import load_iris
iris = load_iris()

# Convert to DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

df.head()


```

```python
# Map numeric to species names
df['species'] = df['species'].map({0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'})
```

# Task 2: Basic data analysis

```python
df.describe() #summary statistics
```

```python
    #grouping by species and calculating mean
df_grouped = df.groupby('species').mean()
print(df_grouped)
```

# Task 3: Data Visualization

```python
import matplotlib.pyplot as plt

# Example of a line chart
plt.figure(figsize=(8, 5))
plt.plot(df.index, df['sepal length (cm)'], label='Sepal Length')
plt.title('Sepal Length Over Index')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.legend()
plt.show()
```

```python
#Bar chart

df_grouped["petal length (cm)"].plot(kind='bar', figsize=(8, 5))
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length (cm)')
plt.xticks(rotation=0)
plt.show()
```

From the bar, we can see specie 2(virginica) has the largest petal length compared to the rest of the species

```python
#Histogram
plt.figure(figsize=(8, 5))
plt.hist(df['sepal length (cm)'], bins=10, edgecolor='black',color='skyblue', alpha=0.7)
plt.title("Distribution of sepal width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
#scatter plot of sepal length vs petal length
plt.figure(figsize=(8, 5))
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'], color='orange', alpha=0.7)
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()
```

As sepal length increases, petal length also increases
